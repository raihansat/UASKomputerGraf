```python
dataset = pd.read_csv("Rainfall_dataset.csv")
dataset.head(15)
```

```python
plt.figure(figsize=(10, 5))

x = np.arange(len(dataset["Bulan"]))

Bulan = [str(i + 1) for i in range(len(dataset["Bulan"]))]

width = 0.28

plt.bar(x - 0.28, dataset["1981-1990"], color = "b", width = width, label="1981-1990")
plt.bar(x + 0.00, dataset["1991-2000"], color = "r", width = width, label="1991-2000")
plt.bar(x + 0.28, dataset["2001-2010"], color = "g", width = width, label="2001-2010")

plt.xlabel("Bulan", fontsize=16, weight="bold")
plt.ylabel("Curah Hujan (mm)", fontsize=16, weight="bold")

plt.legend(loc="upper center", ncol=3)
plt.xticks(x, Bulan)
plt.grid(zorder=0, color="gray", linestyle="dashed", linewidth=0.5)
plt.show()
```

```python
dataset = pd.read_csv("suhu_permukaan.csv")
dataset.head(15)
```

```python
dates = pd.date_range(start='2013-07-24', periods=9, freq='3H')
plt.figure(figsize=(10, 5))

labels = ['Asimilasi', 'Observasi', 'Non Asimilasi']
colors = ['b', 'y', 'r']

for label, color in zip(labels, colors):
  plt.plot(dates, dataset[label], marker='o', color=color, label=label)

for d in [20, 21, 22, 23, 24, 25, 26, 27]:
  plt.axhline(y=d, color='gray', linestyle='-', linewidth=0.3)

plt.gca().xaxis.set_major_formatter(dts.DateFormatter('%H:%M UTC'))
plt.gca().xaxis.set_major_locator(dts.AutoDateLocator())

plt.title('Suhu Permukaan', fontsize=16, weight='bold')
plt.xlabel('Waktu', fontsize=14, weight='bold')
plt.ylabel('Celcius', fontsize=14, weight='bold')
plt.legend(loc='lower right')
plt.show()
```
